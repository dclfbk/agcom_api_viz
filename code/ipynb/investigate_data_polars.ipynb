```python
data = data.with_columns((pl.col("name") + " " + pl.col("lastname")).alias("fullname"))
data
```

```python
#divide political groups (var-->political_groups) and politicians (var-->politicians) in an array with names

political_groups = data.filter(pl.col("name") == "political movement").select("lastname").unique().to_series().to_list()

politicians = data.filter(pl.col("name") != "political movement").select("fullname").unique().to_series().to_list()
```

```python
#find all values of all columns (var: channel, program, affiliation, topic) all arrays

channels = data.select('channel').unique().to_series().to_list()
programs = data.select('program').unique().to_series().to_list()
affiliations = data.select('affiliation').unique().to_series().to_list()
topics = data.select('topic').unique().to_series().to_list()
```

```python
#how much a politician talked about all the possible topics (politicianTopics --> json variable {"topic", "minutes", "interventions"})

n = "Silvio Berlusconi"

x = data.filter(pl.col("fullname") == n)
politicianTopics = []

for t in topics:
  y = x.filter(pl.col("topic") == t)
  sum = y.select("duration").sum().to_series().to_list()
  politicianTopics.append({"topic": t, "minutes" : sum, "interventions" : y.shape[0]})


#how much a political group talked about all the possible topics (polGroupTopics --> json variable {"topic", "minutes", "interventions"})

m = "Azione-Italia Viva-Renew Europe"

w = data.filter(pl.col("lastname") == m)
polGroupTopics = []

for t in topics:
  z = w.filter(pl.col("topic") == t)
  sum = z.select("duration").sum().to_series().to_list()
  polGroupTopics.append({"topic": t, "minutes" : sum, "interventions" : z.shape[0]})
```

```python
#how much a politician talked in a specific channel (politicianChannels --> json variable {"channel", "minutes", "interventions"})

n = "Silvio Berlusconi"

x = data.filter(pl.col("fullname") == n)
politicianChannels = []

for c in channels:
  y = x.filter(pl.col("channel") == c)
  sum = y.select("duration").sum().to_series().to_list()
  politicianChannels.append({"channel": c, "minutes" : sum, "interventions" : y.shape[0]})


#how much a political group talked about all the possible topics (polGroupTopics --> json variable {"topic", "minutes", "interventions"})

m = "Azione-Italia Viva-Renew Europe"

w = data.filter(pl.col("lastname") == m)
polGroupChannels = []

for c in channels:
  z = w.filter(pl.col("channel") == c)
  sum = z.select("duration").sum().to_series().to_list()
  polGroupChannels.append({"channel": c, "minutes" : sum, "interventions" : z.shape[0]})
```

```python
#how many politicians participate (have participated) in every political group (politiciansAffiliation --> json variable {"affiliation", "politicians"[] })

politiciansAffiliation = []
y = data.filter(pl.col("name") != "political movement")

for a in affiliations:
  x = y.filter(pl.col("affiliation") == a)
  politiciansAffiliation.append({"affiliation": a, "politicians": x.select('fullname').unique().to_series().to_list()})
```

```python
#how much a political group has intervened in tv (interventionsPoliticalGroup --> json variable {"affiliation", "interventions", "minutes"})

interventionsPoliticalGroup = []

for a in affiliations:
  x = data.filter(pl.col("affiliation") == a)
  sum = x.select("duration").sum().to_series().to_list()
  interventionsPoliticalGroup.append({"affiliation": a, "interventions": x.shape[0], "minutes": sum})
```

```python
#how many political groups a politician have participated in (politicianPoliticalGroups[])

n = "Matteo Renzi"

x = data.filter(pl.col("fullname") == n)

politicianPoliticalGroups = x.select('affiliation').unique().to_series().to_list()
```

```python
name = "Silvio Berlusconi"
politicians = data.filter(pl.col("name") != "political movement")

politician_channels = politicians.filter(pl.col('fullname') == name)
x = []

for c in channels:
    singleChannel = politician_channels.filter(pl.col('channel') == c)
    programsChannel = singleChannel.select('program').unique().to_series().to_list()
    arr = []
    for p in programsChannel:
        temp2 = singleChannel.filter(pl.col('program') == p)
        p1 = temp2.select('program').unique().to_series().to_list()
        total = temp2.select('duration').sum().to_series().to_list()
        arr.append({'program': p1[0], 'minutes': total[0]})
    x.append({'channel': singleChannel.select('channel').unique().to_series().to_list()[0], 'programs': arr})

final_list = {'politician': name, 'channels': x}
```

```python
name = "Silvio Berlusconi"
channel = "RETE 4"


min_date = "2019/01/01"
max_date = "2022/01/01"
min_year = int(min_date.split('/')[0])
max_year = int(max_date.split('/')[0])

all_programs_channel = data.filter(pl.col('channel') == channel)
programs_channel = all_programs_channel.select('program').unique().to_series().to_list()
programs_channel.sort()

final_programs = []

for prog in programs_channel:
    temp_prgrm = {}
    temp_prgrm["program"] = prog
    temp = all_programs_channel.filter(pl.col('program') == prog)
    temp = temp.filter(pl.col('fullname') == name)
    data = {}
    m_y = 2019
    x_y = 2022
    while m_y != (x_y + 1):
        fy = datetime.strptime(str(min_year), '%Y')
        ly = datetime.strptime(str(min_year + 1), '%Y')
        a = temp.filter(pl.col('day') >= fy)
        a = a.filter(pl.col('day') <= ly)
        data[m_y] = a.select('duration').sum().to_series().to_list()[0]
        m_y += 1
    temp_prgrm["data"] = data
    final_programs.append(temp_prgrm)
```
