```python
tab = duckdb.sql("SELECT *, name || ' ' || lastname AS fullname FROM data")
tab
```

```python
political_groups = duckdb.sql("SELECT DISTINCT lastname FROM tab WHERE name = 'political movement'").fetchall()
politicians = duckdb.sql("SELECT DISTINCT fullname FROM tab WHERE name != 'political movement'").fetchall()
```

```python
channels = duckdb.sql("SELECT DISTINCT channel FROM tab").fetchall()
programs =  duckdb.sql("SELECT DISTINCT program FROM tab").fetchall()
affiliations =  duckdb.sql("SELECT DISTINCT affiliation FROM tab").fetchall()
topics =  duckdb.sql("SELECT DISTINCT topic FROM tab").fetchall()
```

```python
#how much a politician talked about all the possible topics (politicianTopics --> json variable {"topic", "minutes", "interventions"})

n = "Silvio Berlusconi"
politicianTopics = []

#in DuckDB, when a python string is in a query, it is automatically treated like a VARCHAR variable
query = f"SELECT * FROM tab WHERE fullname = '{n}'"
x = duckdb.sql(query)

for t in topics:
    q = f"SELECT * from x WHERE topic = '{t[0]}'"
    y = duckdb.sql(q)
    sum = duckdb.sql("SELECT SUM(duration) FROM y").fetchone()
    interventions = duckdb.sql("SELECT COUNT(*) FROM y").fetchone()
    politicianTopics.append({"topic": t[0], "minutes" : sum[0], "interventions": interventions[0]})


#how much a political group talked in a specific channel (polGroupChannels --> json variable {"channel", "minutes", "interventions"})

m = "Azione-Italia Viva-Renew Europe"
polGroupTopics = []

query = f"SELECT * FROM tab WHERE lastname = '{m}'"
w = duckdb.sql(query)

for t in topics:
    q = f"SELECT * from w WHERE topic = '{t[0]}'"
    z = duckdb.sql(q)
    sum = duckdb.sql("SELECT SUM(duration) FROM z").fetchone()
    interventions = duckdb.sql("SELECT COUNT(*) FROM z").fetchone()
    polGroupTopics.append({"topic": t[0], "minutes" : sum[0], "interventions": interventions[0]})
```
